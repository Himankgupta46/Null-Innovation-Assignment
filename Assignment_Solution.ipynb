```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel(r'D:\Assignment INSAID\NULL inovation\Tweet_NFT.xlsx')
data.head()
```

```python
training_data = data[data['tweet_intent'].isnull() == False] #spliting the data set for training which have the tweet_intent present.
training_data.head()
```

```python
testing_data = data[data['tweet_intent'].isnull() == True] #spliting the data set for testing which have the tweet_intent are not present.
testing_data.head()
```

```python
x_train = training_data['tweet_text']
y_train = training_data['tweet_intent']

x_test = testing_data['tweet_text']
y_test = testing_data['tweet_intent']

from sklearn.feature_extraction.text import CountVectorizer # To convert text data into numerical data in the form of vector.
from sklearn.feature_extraction.text import TfidfTransformer # To convert matrix into the normalized tfid.

count_vect = CountVectorizer() 
X_train_counts = count_vect.fit_transform(x_train) 
X_train_counts.shape

tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape

from sklearn.naive_bayes import MultinomialNB # Use the Multinomial Naive Bayes to predict (best fit for NLP)
clf = MultinomialNB().fit(X_train_tfidf, y_train)

print(clf.predict(count_vect.transform(["I want to buy a NFT"])))
```

```python
def predict_intent(text):                                   # Creating function to represnt the predicted output in the form or data set.
    return clf.predict(count_vect.transform([text]))[0]

testing_data['tweet_intent'] = testing_data[0:10]['tweet_text'].apply(predict_intent)
testing_data.head(10)
```
